```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2-medium"  # We can use other models like "gpt2-medium" or "gpt2-large" for better performance
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()
```

```python
def generate_email(prompt, max_length=200, temperature=0.7):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text using the model with do_sample=True
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        num_return_sequences=1,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,  # Set do_sample to True
    )

    # Decode and return the generated email
    generated_email = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_email
```

```python
prompt = "URGENT! You have won a 1 week FREE membership in our £100,000 Prize Jackpot! Txt the word: CLAIM to No: 81010 T&C www.dbuk.net LCCLTD POBOX 4403LDNW1A7RW18"
generated_email = generate_email(prompt)

print("Generated Email:")
print(generated_email)
```
